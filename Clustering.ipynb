```python
plt.figure(figsize = (15, 10))

bskills = combine.iloc[:, 24:].cumsum().iloc[-1].sort_values()
plt.barh(bskills.index, bskills, color = 'lightcoral', label = 'Business Skills')

pskills = combine.iloc[:, :24].cumsum().iloc[-1].sort_values()
plt.barh(pskills.index, pskills, color = 'slategray', label = 'Programing Skills')

plt.legend(fontsize = 'large')
plt.title('Programing Skills and Business Skills', fontsize = 20)
plt.show()
```

```python
import circlify

# compute circle positions:
circles = circlify.circlify(
    pskills[2:].to_list(), 
    show_enclosure=False, 
    target_enclosure = circlify.Circle(x = 0, y = 0, r = 1)
)

fig, ax = plt.subplots(figsize = (10, 10))

ax.axis('off')
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

labels = pskills[2:].index

for circle, label in zip(circles, labels):
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r, alpha = 0.2, linewidth = 2, color = 'slategray'))
    plt.annotate(
          label, 
          (x,y ) ,
          va = 'center',
          ha = 'center'
     )
```

### Hierarchical Clustering

```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import zeros as sci_zeros
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as sch
from scipy import zeros as sci_zeros
from scipy.spatial.distance import euclidean
from sklearn.preprocessing import StandardScaler
```

```python
num_skills = len(combine.columns)
Skill_occur = sci_zeros([num_skills, num_skills])

# Count the time two skills appears at the same time
for i in range(num_skills):
    for j in range(num_skills):
        Skill1 = combine.iloc[:, i] == 1
        Skill2 = combine.iloc[:, j] == 1
        Skill_occur[i,j] = len(combine[Skill1 & Skill2])
```

```python
fig= plt.figure(figsize=(24,12))
ax1 = fig.add_axes([0.3,0.3,0.6,0.7])
dendrogram = hierarchy.dendrogram(hierarchy.linkage(Skill_occur,method="ward"),orientation="right")
ax1 = ax1.set_yticklabels(combine.columns)
```

```python
from sklearn.preprocessing import StandardScaler
combine_1 = combine.T
combine_1.set_index(combine.columns)
scaler = StandardScaler()
scaler.fit(combine_1)
combine_1= scaler.transform(combine_1)
```

```python
#create dendrogram
fig= plt.figure(figsize=(24,12))
ax1 = fig.add_axes([0.1,0.1,0.4,0.6])
dendrogram = hierarchy.dendrogram(hierarchy.linkage(combine_1,method="ward"),orientation="right")
ax1 = ax1.set_yticklabels(combine.columns)
```

```python
combine_1 = combine.T
combine_1.set_index(combine.columns)
plt.figure(figsize=(16, 10))
linkage = hierarchy.linkage(combine_1, 'ward')
hierarchy.dendrogram(linkage)
plt.title('Dendrograms Combined')
plt.xticks(combine_1.index)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
linkage = hierarchy.linkage(indeed.iloc[:, 7:], 'ward')
hierarchy.dendrogram(linkage)
plt.title('Dendrograms Indeed')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
linkage = hierarchy.linkage(sh.iloc[:, 7:], 'ward')
hierarchy.dendrogram(linkage)
plt.title('Dendrograms Simplyhired')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
linkage = hierarchy.linkage(pjf.iloc[:, 7:], 'ward')
hierarchy.dendrogram(linkage)
plt.title('Dendrograms PostJobFree')
plt.show()
```
